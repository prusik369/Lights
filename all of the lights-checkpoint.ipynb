```python
def data_preparation(sales):
    X.LotArea.where(X.LotArea < 80000, 80000, inplace=True)
    X.MasVnrType = X.MasVnrType.fillna("None")
    X.MasVnrArea = X.MasVnrArea.fillna(0)
    X['2ndFlrSF'] = X['2ndFlrSF'].astype('bool').astype(int)
    X['WoodDeckSF'] = X['WoodDeckSF'].astype('bool').astype(int)
    X['OpenPorchSF'] = X['OpenPorchSF'].astype('bool').astype(int)
    X['EnclosedPorch'] = X['EnclosedPorch'].astype('bool').astype(int)
    X['3SsnPorch'] = X['3SsnPorch'].astype('bool').astype(int)
    X['ScreenPorch'] = X['ScreenPorch'].astype('bool').astype(int)
    X["PorchSum"] = X['WoodDeckSF'] + X['OpenPorchSF'] + X['EnclosedPorch'] + X['3SsnPorch'] + X['ScreenPorch']
    ordinals = {"Ex" : 5, "Gd": 4, "TA" : 3, "Fa": 2, "Po" : 1, "NA" : 0}
    X.ExterQual = X.ExterQual.map(ordinals)
    X.ExterCond = X.ExterCond.map(ordinals)
    X.BsmtQual = X.BsmtQual.map(ordinals)
    X.BsmtCond = X.BsmtCond.map(ordinals)
    X.HeatingQC = X.HeatingQC.map(ordinals)
    X.KitchenQual = X.KitchenQual.map(ordinals)
    X.FireplaceQu = X.FireplaceQu.map(ordinals)
    X.GarageQual = X.GarageQual.map(ordinals)
    X.GarageCond = X.GarageCond.map(ordinals)
    X.GarageCond = X.GarageCond.map(ordinals)
    X = X.drop(columns=['Id',
                        'Street',
                        'YearBuilt',
                        '1stFlrSF',
                        'LowQualFinSF',
                        'GarageArea',
                        'MoSold',
                        'YrSold',
                        'WoodDeckSF',
                        'OpenPorchSF',
                        'EnclosedPorch',
                        '3SsnPorch',
                        'ScreenPorch',
                        'OverallCond',
                        'PoolArea',
                        'PoolQC',
                        'MiscVal']
                               ,inplace = True)
    return X
```
